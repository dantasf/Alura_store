```python
# Alura Store - An√°lise de Vendas
# Autor: Paulo R. Dantas
# Objetivo: Ajudar o Sr. Jo√£o a decidir qual loja vender com base nos dados

import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados (arquivos devem estar no mesmo diret√≥rio do notebook)
loja1 = pd.read_csv("loja_1.csv")
loja2 = pd.read_csv("loja_2.csv")
loja3 = pd.read_csv("loja_3.csv")
loja4 = pd.read_csv("loja_4.csv")

# Organizar em dicion√°rio
lojas = {
    "Loja 1": loja1,
    "Loja 2": loja2,
    "Loja 3": loja3,
    "Loja 4": loja4,
}

# ---------- AN√ÅLISES ---------- #

# 1. Faturamento total por loja
faturamentos = {nome: df["Pre√ßo"].sum() for nome, df in lojas.items()}
faturamento_df = pd.DataFrame.from_dict(faturamentos, orient="index", columns=["Faturamento"])
print("Faturamento total por loja:")
print(faturamento_df, "\n")

# 2. Vendas por categoria
print("Vendas por categoria:")
for nome, df in lojas.items():
    print(f"\n{nome}:\n", df["Categoria do Produto"].value_counts())

# 3. M√©dia de avalia√ß√£o por loja
avaliacoes = {nome: df["Avalia√ß√£o da compra"].mean().round(2) for nome, df in lojas.items()}
avaliacao_df = pd.DataFrame.from_dict(avaliacoes, orient="index", columns=["M√©dia Avalia√ß√£o"])
print("\nM√©dia de avalia√ß√µes por loja:")
print(avaliacao_df, "\n")

# 4. Produtos mais e menos vendidos
print("Produtos mais e menos vendidos:")
for nome, df in lojas.items():
    contagem = df["Produto"].value_counts()
    print(f"\n{nome}:")
    print("Mais vendido:", contagem.idxmax(), "-", contagem.max(), "vendas")
    print("Menos vendido:", contagem.idxmin(), "-", contagem.min(), "vendas")

# 5. Frete m√©dio por loja
fretes = {nome: df["Frete"].mean().round(2) for nome, df in lojas.items()}
frete_df = pd.DataFrame.from_dict(fretes, orient="index", columns=["Frete M√©dio"])
print("\nFrete m√©dio por loja:")
print(frete_df)

# ---------- VISUALIZA√á√ïES ---------- #

# Gr√°fico 1 - Faturamento por loja
faturamento_df.plot(kind="bar", legend=False, title="Faturamento por Loja", ylabel="R$", color="orange", figsize=(8,5))
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# Gr√°fico 2 - M√©dia de avalia√ß√µes
avaliacao_df.plot(kind="barh", legend=False, title="M√©dia de Avalia√ß√£o por Loja", xlim=(0,5), color="green", figsize=(8,4))
plt.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# Gr√°fico 3 - Frete m√©dio
frete_df.plot(kind="line", marker="o", title="Frete M√©dio por Loja", ylabel="R$", color="purple", figsize=(8,5))
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# ---------- RELAT√ìRIO FINAL ---------- #

print("\n--- RECOMENDA√á√ÉO FINAL ---\n")

recomendacao = """
Ap√≥s an√°lise de desempenho das quatro lojas da Alura Store, conclu√≠mos o seguinte:

- A **Loja 3** apresenta o menor faturamento total, sendo a menos lucrativa.
- Ela tamb√©m tem uma das piores m√©dias de avalia√ß√£o e n√£o se destaca nas categorias mais vendidas.
- Seu frete m√©dio n√£o √© competitivo em rela√ß√£o √†s outras.
- Os produtos mais vendidos nela s√£o comuns √†s outras lojas, sem diferencial.

üìâ **Recomenda√ß√£o:** o Senhor Jo√£o deve vender a **Loja 3**, pois ela √© a que apresenta o pior desempenho geral nos indicadores analisados.
"""

print(recomendacao)
```
